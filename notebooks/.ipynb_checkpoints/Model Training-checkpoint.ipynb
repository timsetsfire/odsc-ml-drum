#### Import Libraries

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import pickle
```

#### Import Data

```python
df = pd.read_excel('../data/boston_housing.xlsx')
df.head()
```

```python
X = df.drop('medv', axis=1)
y = df['medv']
```

#### Train Model

```python
rf = RandomForestRegressor(n_estimators = 20)
rf.fit(X,y)
```

#### Save Model

```python
with open('../src/custom_model/rf.pkl', 'wb') as pkl:
    pickle.dump(rf, pkl)
```
